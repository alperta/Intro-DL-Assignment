# Importing libraries and packages
We first import all the libraries and packages we'll need for this assignment

```python
#!pip install numpy pandas tensorflow keras sklearn tqdm matplotlib

import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

# Importing the data set
Getting the data from the net

```python
# putting the data in a Pandas DataFrame
data_df = pd.read_csv('http://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
data_df.head()
```

# A. Build a baseline model
## Data Wrangling

```python
# checking if data contains null values
data_df.isnull().sum()
```

```python
# we remove the Strength column from the data and store it in y as targets
y = data_df.pop('Strength')
y.head()
```

```python
# the remaining data is X, the predictors used to train the model
X = data_df
X.head()
```

```python
# we check the sets have the same number of rows
print(X.shape, y.shape)
```

## Create and evaluate model
Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the train_test_splithelper function from Scikit-learn.

```python
# we split the predictors (X) and targets (y) in train and test sets with 30% of the data for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

#checking the size of training and testing sets for both X and y
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

Train the model on the training data using 50 epochs.

Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.

Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.

Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.

```python
# creating a neural network with 1 hidden layer of 10 nodes, and a ReLU activation function
# using the adam optimizer and the mean squared error as the loss function
def create_model_1():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(X_train.shape[1],))) # input shape is the number of columns of X_train
    model.add(Dense(1))

    # compiling the model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# creating a list to hold all mean squared errors
mse_list_model_a = []

# training the model 50 times and getting the mse for each iteration, storing the values in the list
for i in range(50):
    # creating the model
    print(f"Model A epoch {i+1} of 50")
    model_a = create_model_1()
    
    #fitting the model
    model_a.fit(X_train, y_train, validation_split=0.3, epochs=50, verbose=0, batch_size=64)
    
    # using the model to predict values
    y_hat_a = model_a.predict(X_test)
    
    # append mse between predicted data and actual data
    mse_list_model_a.append(mean_squared_error(y_test, y_hat_a))   
```

## Compute MSE
Report the mean and the standard deviation of the mean squared errors

```python
def get_mean_and_std_from_mse_list(name, mse_list):
    return pd.DataFrame({"Model":name, "Mean": np.asarray(mse_list).mean(), "St. Dev.": np.asarray(mse_list).std()}, index=[0])

results = get_mean_and_std_from_mse_list('A', mse_list_model_a)
results
```

# B. Normalize the data
Repeat Part A but use a normalized version of the data. Recall that one way to normalize the data is by subtracting the mean from the individual predictors and dividing by the standard deviation.

```python
# we normalize the predictors
X_norm = (X_train - X_train.mean()) / X_train.std()
X_norm.head()
```

```python
#we normalize the test data as well
X_norm_test = (X_test - X_test.mean()) / X_test.std()

# same as part A, but with the normalized data
# creating a list to hold all mean squared errors
mse_list_model_b = []

# training the model 50 times and getting the mse for each iteration, storing the values in the list
for i in range(50):
    # creating the model
    print(f"Model B epoch {i+1} of 50")
    model_b = create_model_1()
    
    #fitting the model with the normalized data
    model_b.fit(X_norm, y_train, validation_split=0.3, epochs=50, verbose=0, batch_size=64)
    
    # using the model to predict values
    y_hat_b = model_b.predict(X_norm_test)
    
    # append mse between predicted data and actual data
    mse_list_model_b.append(mean_squared_error(y_test, y_hat_b))
```

## Compute MSE
Report the mean and the standard deviation of the mean squared errors with normalized predictors

```python
results = pd.concat([results, get_mean_and_std_from_mse_list('B', mse_list_model_b)])
results
```

**The mean of model B is marginaly better with model A, but the std dev is much better. The model seems to have a lot more confidence in the mean value.**

# C. Increate the number of epochs
Repeat Part B but use 100 epochs this time for training.

```python
# creating a list to hold all mean squared errors
mse_list_model_c = []

# training the model 100 times and getting the mse for each iteration, storing the values in the list
for i in range(100): # tqdm is a library to get a nice progression bar
    # creating the model
    print(f"Model C epoch {i+1} of 100")
    model_c = create_model_1()
    
    #fitting the model with the normalized data
    model_c.fit(X_norm, y_train, validation_split=0.3, epochs=50, verbose=0, batch_size=64)
    
    # using the model to predict values
    y_hat_c = model_c.predict(X_norm_test)
    
    # append mse between predicted data and actual data
    mse_list_model_c.append(mean_squared_error(y_test, y_hat_c))
```

## Compute MSE
Report the mean and the standard deviation of the mean squared errors with 100 epochs

```python
results = pd.concat([results, get_mean_and_std_from_mse_list('C', mse_list_model_c)])
results
```

**In this case, increasing the number of epochs in model C doesn't seem to make any meaningful difference with model B**

# D. Increase the number of hidden layers
Repeat part B but use a neural network with 3 hidden layers, each of 10 nodes and ReLU activation function.



```python
# creating a neural network with 3 hidden layers of 10 nodes, and a ReLU activation function
# using the adam optimizer and the mean squared error as the loss function
def create_model_2():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(X_train.shape[1],))) # input shape is the number of columns of X_train
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))

    # compiling the model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# creating a list to hold all mean squared errors
mse_list_model_d = []

# training the model 50 times and getting the mse for each iteration, storing the values in the list
for i in range(50):
    # creating the model
    print(f"Model D epoch {i+1} of 50")
    model_d = create_model_2()
    
    #fitting the model with the normalized data
    model_d.fit(X_norm, y_train, validation_split=0.3, epochs=50, verbose=0, batch_size=64)
    
    # using the model to predict values
    y_hat_d = model_d.predict(X_norm_test)
    
    # append mse between predicted data and actual data
    mse_list_model_d.append(mean_squared_error(y_test, y_hat_d))
```

```python
results = pd.concat([results, get_mean_and_std_from_mse_list('D', mse_list_model_d)])
results
```

**Model D has both a (way) smaller mean and st. dev. error, which tends to prove that more complex neural networks perform better**
